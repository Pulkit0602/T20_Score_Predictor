```python
req_data["Over"] = req_data["delivery"].apply(lambda x:str(x).split(".")[0])
req_data["Balls"] = req_data["delivery"].apply(lambda x:str(x).split(".")[1])
```

```python
req_data["Over"] = req_data["Over"].astype("int")
req_data["Balls"] = req_data["Balls"].astype("int")
```

```python
req_data
```

```python
req_data["Balls_Played"] = req_data["Over"]* 6 + req_data["Balls"]
req_data["Balls_Left"] = 120 - req_data["Balls_Played"]
```

***HANDLING PLAYER OUT***

```python
req_data["player_out"]  = req_data["player_out"].apply(lambda x: 0 if x == 0 else 1)
req_data["player_out"] = req_data["player_out"].astype("int")
req_data["player_out"] = req_data.groupby("match_id").cumsum()["player_out"]
```

```python
req_data["player_left"] = 10 - req_data["player_out"]
```

***HANDLING NEGATIVE VALUE IN Balls_Left***


```python
req_data["Balls_Left"]  = req_data["Balls_Left"].apply(lambda x: 0 if x <0 else x)
```

***Introducig Run Rate***

```python
req_data["Run_Rate"]  = round((req_data['current_score']*6)/req_data['Balls_Played'],2)
```

***Last-Five Overs Runs***

```python
groups = req_data.groupby('match_id')

match_ids = req_data['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
req_data["last_five"] = last_five
```

### CREATING MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

***TRAIN - TEST SPLIT***

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
ohe = OneHotEncoder()
ohe.fit(X[["batting_team" , "bowling_team" , "city" ]])
```

```python
ohe.categories_
```

```python
categorical_transformer = OneHotEncoder(categories = ohe.categories_ ,sparse=False,drop='first')
numerical_transformer = StandardScaler()
numerical_Column = ["current_score" , "Balls_Left" , "player_left" , "Run_Rate" , "last_five"]
categorial_Column = ["batting_team" , "bowling_team" , "city" ]

#Column Transformer
column_transformer = ColumnTransformer(transformers = [("cat" ,categorical_transformer , categorial_Column) , 
                                                       ("num" , numerical_transformer , numerical_Column) ] , 
                                                          remainder = "passthrough")
```

```python
model1 = Pipeline(steps=[
    ('step1',column_transformer),
    ('step2',RandomForestRegressor())])
```

```python
model1.fit(X_train , y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
